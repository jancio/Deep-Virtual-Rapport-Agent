```python
#######################################################################################################################
# Plot distributions of gaze_angle_x, gaze_angle_y extracted by OpenFace
#######################################################################################################################

import glob
import time
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

input_features_dir = f'/home/ICT2000/jondras/datasets/mimicry/hgd_annotated_features_perframe'

start_time = time.time()
cnt = 0
for feature_file in sorted(glob.glob(input_features_dir + '/*.csv')):
    cnt += 1
    df = pd.read_csv(feature_file)
    
    # Angles are converted from radians to degrees
    gaze_angle_x = df[' gaze_angle_x'] * 180 / np.pi
    gaze_angle_y = df[' gaze_angle_y'] * 180 / np.pi
    
    bins_range = 100
    print(f'Mean: {np.mean(gaze_angle_x)}\t Median: {np.median(gaze_angle_x)}\t STD: {np.std(gaze_angle_x)}')
    print(f'Min: {np.min(gaze_angle_x)}\t Max: {np.max(gaze_angle_x)}')
    fig = plt.figure()
    plt.title(feature_file.split('/')[-1])
    plt.hist(gaze_angle_x, bins=bins_range)
    plt.xlabel('Gaze angle X (deg)')
    plt.ylabel('Count')
    plt.show()
    fig.clear()
    plt.close(fig)
    
    print(f'Mean: {np.mean(gaze_angle_y)}\t Median: {np.median(gaze_angle_y)}\t STD: {np.std(gaze_angle_y)}')
    print(f'Min: {np.min(gaze_angle_y)}\t Max: {np.max(gaze_angle_y)}')
    fig = plt.figure()
    plt.title(feature_file.split('/')[-1])
    plt.hist(gaze_angle_y, bins=bins_range)
    plt.xlabel('Gaze angle Y (deg)')
    plt.ylabel('Count')
    plt.show()
    fig.clear()
    plt.close(fig)
    
    print(f'=================================================================================\n')
    
print(f'Processed {cnt} feature files.')
```
