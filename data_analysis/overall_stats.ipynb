```python
#######################################################################################################################
# Get histograms of head gesture durations from vra1, sewa, nvb, and hatice2010 datasets (i.e., 4comb dataset)
# (ccdb is not included)
#######################################################################################################################

import glob
import numpy as np
import matplotlib.pyplot as plt


gestures = ['nod', 'shake', 'tilt']
bins_range = 100

for g in gestures:
    print(f'Overall {g} stats:')
    durations = []
    for filename in glob.glob(f'./gesture_durations/{g}_*.npz'):
        if 'ccdb' not in filename:
            durations.extend(np.load(filename)['durations'])
    print(f'\t Mean: {np.mean(durations)}\t Median: {np.median(durations)}')
    print(f'\t Min: {np.min(durations)}\t Max: {np.max(durations)}\n')
    
    plt.figure()
    plt.title(g)
    plt.hist(durations, bins=bins_range, label=f'total {g}: {len(durations)}')
    plt.xlabel('Duration (sec)')
    plt.ylabel('Count')
    plt.legend()
    plt.show()
```
