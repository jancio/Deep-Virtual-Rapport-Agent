```python
#######################################################################################################################
# Project: Deep Virtual Rapport Agent (rapport model)
#
#     Jan Ondras (jo951030@gmail.com)
#     Institute for Creative Technologies, University of Southern California
#     April-October 2019
#
#######################################################################################################################
# Separate audio channels of the two subjects from each stereo audio recording
#######################################################################################################################

import glob
import soundfile as sf

# Path to save mono audio files
mono_audio_files_path_prefix = '/home/ICT2000/jondras/dvra_datasets/mimicry/audio/audio_separated_48kHz'
# Path to extracted features to get correct position P1/P2 for each subject id
features_path_prefix = '/home/ICT2000/jondras/dvra_datasets/mimicry/vision_features/original_openface_features'

# Iterate over original stereo audio files
for i, stereo_audio_file in enumerate(sorted(glob.glob('/home/ICT2000/jondras/dvra_datasets/mimicry/original_data/sessid_*/Sessions/*/*_TrimmedAudio_c1_c2.wav'))):
    sessid = stereo_audio_file.split('/')[-4].split('_')[-1]
    # Skip incomplete session 39
    if sessid == '39':
        continue
    print(stereo_audio_file, sessid)
    
    # Keep original sample rate
    stereo_data, samplerate = sf.read(stereo_audio_file)
    # print(stereo_data.shape)
    P1_mono_data = stereo_data[:, 0]
    P2_mono_data = stereo_data[:, 1]
      
    P1_sid = glob.glob(f'{features_path_prefix}/sessid_{sessid}_P1_*')[0].split('/')[-1].split('_')[-1][:-4]
    P2_sid = glob.glob(f'{features_path_prefix}/sessid_{sessid}_P2_*')[0].split('/')[-1].split('_')[-1][:-4]
    print(P1_sid, P2_sid)
    
    sf.write(f'{mono_audio_files_path_prefix}/sessid_{sessid}_P1_sid_{int(P1_sid):02d}.wav', P1_mono_data, samplerate)
    sf.write(f'{mono_audio_files_path_prefix}/sessid_{sessid}_P2_sid_{int(P2_sid):02d}.wav', P2_mono_data, samplerate)
#     break

print(f'Audio separated for {i} sessions.')
```
