```python
#######################################################################################################################
# Unify listener nod annotation filenames: SES{id}.eaf
#######################################################################################################################


from os import rename
import glob

for old_file_path in sorted(glob.glob('/home/ICT2000/jondras/dvra_datasets/vra1/listener_head_gesture_annotations/*.L.nod.eaf')):
    split_file_path = old_file_path.split('/')
    root_start_idx = split_file_path[-1].find('SES')    
    new_file_path = '/'.join(split_file_path[:-1]) + '/' + split_file_path[-1][root_start_idx:].split('.')[0] + '.eaf'
    rename(old_file_path, new_file_path)
    print(f'{old_file_path}\n\t renamed to: {new_file_path}')
```

```python
#######################################################################################################################
# Unify listener video filenames: SES{id}.mp4
#######################################################################################################################


from os import rename
import glob

for old_file_path in sorted(glob.glob('/home/ICT2000/jondras/dvra_datasets/vra1/listener_videos/*.mp4')):
    split_file_path = old_file_path.split('/')
    root_start_idx = split_file_path[-1].find('SES')    
    new_file_path = '/'.join(split_file_path[:-1]) + '/' + split_file_path[-1][root_start_idx:].split('.')[0] + '.mp4'
    rename(old_file_path, new_file_path)
    print(f'{old_file_path}\n\t renamed to: {new_file_path}')
```
